```python
def get_iso_week_start(date_series):
    date_series = pd.to_datetime(date_series)
    weekday = date_series.dt.isoweekday
    return date_series - pd.to_timedelta(weekday - 1, unit='D')
```

```python
import pandas as pd

# Read all lines from the file
with open('FBDaily_data_Pt2', 'r') as file:
    lines = [line.strip() for line in file.readlines()]

# First 11 lines become column names
column_names = lines[:11]

# Remaining lines are the data
data_lines = lines[11:]

# Group remaining lines into chunks of 11
rows = []
for i in range(0, len(data_lines), 11):
    chunk = data_lines[i:i+11]
    if len(chunk) == 11:  # Only add complete rows
        rows.append(chunk)

# Create DataFrame
df = pd.DataFrame(rows, columns=column_names)
print(df['Date'].max())
print(df['Date'].min())
fb_df = df.copy()

```
